# 1. Definitions



(1) What is SQL?

Structured query language (SQL) is a domain-specific, programming language used for storing and processing information in a relational database. It's a declarative programming language. We can use SQL to store, update, delete, search, and retrieve information from the database. SQL has been standardized by The American National Standards Institute (ANSI) and the International Standards Organization (ISO). 

(2) What is a database?

A database is a repository or an organized collection of data. However, the term database also typically refers to how that data is accessed, manipulated or managed.

(3) What is DBMS, and what types of DBMS do you know?

A DBMS (Database Management Software) is software that gives users and applications the ability to define, create, query and administer a database. It translates requests for data manipulation between the clients and the database itself. 

Types of DBMS of which I'm aware:

* Relational, e.g. Oracle, MySQL, PostgreSQL

* Object, e.g.  Microsoft Access, Firebird, InterBase

* NoSql / Non-Relational, e.g.  MongoDB, CouchDB

    * Graph, e.g. Neo4j

    * Key-value, e.g. Amazon DynamoDB
    
    * Triple/rdf store

    * Document

* NewSQL, e.g.  Google Spanner, Apache Ignite

* hybrids of the above

(4) What aggregate functions do you know?

- avg(X): Calculates the average (mean) of values in X.

- count(X) / count(*): Returns the number of rows where X is not null / Returns the total number of rows.

- group_concat(X) / group_concat(X,Y): Concatenates the non-null values of X within each group / Concatenates the 
non-null values of X separated by Y within each group.

- max(X): Returns the maximum value of X.

- min(X): Returns the minimum value of X.

- sum(X): Calculates the sum of values in X.

- mode(X): Returns the most frequent value in X.

- median(X): Calculates the median value of X.

- total(X): Returns the sum of all values in X.

where X is an aggregate

(5) What is the difference between the HAVING and WHERE statements?

The WHERE clause is used to filter rows before any aggregation is performed. It is applied to individual rows in the result set. It is used with the SELECT, UPDATE, DELETE, and INSERT statements to specify conditions that must be met for the rows to be included in the result set.

The HAVING clause is exclusively used to filter the result set after the aggregation is performed, typically when we're using aggregate functions like COUNT, SUM, AVG, etc. It is applied to aggregated data, such as the result of a GROUP BY clause. It is used to specify conditions that must be met by groups of rows, rather than individual rows.

1) Write a SQL query to separate the names of employees and their leaders by the string 'works for'. Return result like "BLAZE works for KAYLING". Print out the results.


```python
query = '''SELECT e1.emp_name || ' works for ' || e2.emp_name
FROM employees e1 JOIN employees e2 ON e1.leader_id = e2.emp_id ;'''
allrowsSelect(conn, query)
```

8) Write a query to find the names and emp_ID of the employees who are leaders. Sort the table by emp_ID in ascending order. Print out the result and save the result in another table "leaders".

```python
query = '''SELECT e2.emp_name, e2.emp_id
FROM employees e1 JOIN employees e2 ON e1.leader_id = e2.emp_id GROUP BY e1.leader_id ORDER BY e2.emp_id;'''
allrowsSelect(conn, query)
```

```python
query = '''CREATE TABLE leaders (emp_name TEXT, emp_id INTEGER);'''
execSQL(conn, query)
```

```python
query = '''INSERT INTO leaders (emp_name, emp_id)
SELECT e2.emp_name, e2.emp_id
FROM employees e1 
JOIN employees e2 ON e1.leader_id = e2.emp_id 
GROUP BY e2.emp_name, e2.emp_id 
ORDER BY e2.emp_id;'''
execSQL(conn, query)
```

9) Write a query to find the jobs having an average salary exceed 2700. Sort the table in descending order by average price. Only output a table with the job types and their average salary.

```python
query = '''SELECT job_name, AVG(total_income) AS average_salary from employees GROUP BY job_name HAVING average_salary > 2700 ORDER BY average_salary desc;'''
allrowsSelect(conn, query)
```
